# Definitions
* __DataFrame__: Its just table with rows and columns
* __Series__: Each of the Columns from DataFrame is called panda series
* Important functions
    * read_csv : read , seperated file
    * read_table: read /t seperated file
    * head(): check top 5 rows
    * type: Data Type
    * describe(): return statical info like mean etc
    * shape: provides the shape of dataframe. number of rows, number of cols
    * dtypes: datatypes of each column

```python
import pandas as pd
```

# DataFrame, read_table

```python
#easy
orders = pd.read_table('http://bit.ly/chiporders') #default values of read_table is /t seperated and first column is header
print(type(orders))
print(orders.shape)
orders.head()
```

__Read only few columns and few rows. it will make faster__

```python
fewCols=['quantity','item_name']
ordersFewColumns = pd.read_table('http://bit.ly/chiporders',usecols=fewCols,nrows=5)
print(type(ordersFewColumns))
print(ordersFewColumns.shape)
ordersFewColumns.head()
```

```python
__Remove non-numeric columns__
```

```python
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
print(drinks.dtypes)

import numpy as np
drinks.select_dtypes(include=(np.number)).dtypes
```

```python
#bit difficult
users=pd.read_table('http://bit.ly/movieusers',sep='|',header=None)
print("Without Headers")
users.head()
```

```python
#Above no column names, below we are specifying colnames
headers=['Id','Age','Gender','Profession','ZipCode']
usersWithHeaders=pd.read_table('http://bit.ly/movieusers',sep='|',header=None,names=headers)
print("With Headers")
usersWithHeaders.head()
```

# Change column names

# Before

```python
ufo = pd.read_csv('http://bit.ly/uforeports') 
ufo.head()
ufo.columns
```

__Method 1__

```python
ufo = pd.read_csv('http://bit.ly/uforeports') 
ufo.head()
ufo.rename(columns={'Colors Reported':'ColorsReported'},inplace=True)
ufo.columns
```

__Method 2__

```python
ufo = pd.read_csv('http://bit.ly/uforeports') 
ufo.head()
cols=['City','ColorsReported','ShapeReported','State','Time']
ufo.columns=cols
ufo.columns
```

__Method 3__

```python
ufo = pd.read_csv('http://bit.ly/uforeports') 
ufo.head()
ufo.columns=ufo.columns.str.replace(' ','_')
ufo.columns
```

# Remove Columns

```python
ufo = pd.read_csv('http://bit.ly/uforeports') 
ufo.head()
ufo.columns
```

__Method 1__

```python
ufo.drop('Colors Reported',axis=1,inplace=True)
ufo.head()
```

__Method 2__

```python
ufo.drop(['City','State'],axis=1,inplace=True)
ufo.head()
```

# Remove Rows

```python
ufo = pd.read_csv('http://bit.ly/uforeports') 
ufo.head()
ufo.columns
ufo.drop([0,1],axis=0,inplace=True)
ufo.head()
```
