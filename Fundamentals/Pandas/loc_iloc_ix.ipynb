```python
import pandas as pd
```

```python
data= pd.read_csv('http://bit.ly/uforeports')
```

```python
data.head()
```

# Loc
* Pull rows and columns

```python
# 1st row,All columns , : is used for all
data.loc[0,:]
```

```python
#1,2,4 rows, all columns
data.loc[[1,2,4],:]
```

```python
# continous rows
data.loc[10:13,:]
```

```python
#filtering
data.loc[data['City']=='Oakland','State']
```

# iloc vs loc vs ix
* iloc is to fetch by integer whereas loc is fetch by label. ix- both can be specified. 
* iloc-last value is exclusive but loc-last value is inclusive

```python
data.loc[0:3,:]
```

```python
data.iloc[0:3,:]
```

```python
data.loc[2,'City':'State']
```

# Map

```python
train=pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
#map: replace as dictionary

train['SexNum']=train['Sex'].map({'male':0,'female':1})
train.loc[0:5,['Sex','SexNum']]
```

```python
#apply: apply to whole column
train['NameLength']=train.Name.apply(len)
train.head()
```

```python
#applymap: apply map to whole dataframe
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
drinks.loc[:,['beer_servings','spirit_servings']].applymap(float)
```
